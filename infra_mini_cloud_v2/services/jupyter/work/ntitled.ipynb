```python
import os
from pyspark.sql import SparkSession

os.environ['PYSPARK_SUBMIT_ARGS'] = (
    '--packages '
    'org.apache.iceberg:iceberg-spark-runtime-3.5_2.12:1.5.2,'
    'org.apache.hadoop:hadoop-client-api:3.3.4,'
    'org.apache.hadoop:hadoop-client-runtime:3.3.4,'
    'org.apache.hadoop:hadoop-aws:3.3.4,'
    'com.amazonaws:aws-java-sdk-bundle:1.12.262 '
    'pyspark-shell'
)

spark = SparkSession.builder \
    .appName("Iceberg Hadoop Catalog no Ceph") \
    .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions") \
    .config("spark.sql.catalog.ceph_cat", "org.apache.iceberg.spark.SparkCatalog") \
    .config("spark.sql.catalog.ceph_cat.type", "hadoop") \
    .config("spark.sql.catalog.ceph_cat.warehouse", "s3a://raw/") \
    .config("spark.hadoop.fs.s3a.endpoint", "http://192.168.100.10:8080") \
    .config("spark.hadoop.fs.s3a.access.key", "myaccesskey") \
    .config("spark.hadoop.fs.s3a.secret.key", "mysecretkey") \
    .config("spark.hadoop.fs.s3a.path.style.access", "true") \
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .config("spark.hadoop.fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider") \
    .config("spark.hadoop.fs.s3a.attempts.maximum", "1") \
    .getOrCreate()

spark.sql("CREATE NAMESPACE IF NOT EXISTS ceph_cat.default")

df = spark.createDataFrame([
    (1, "Alice", 30),
    (2, "Bob", 25),
    (3, "Carlos", 35)
], ["id", "nome", "idade"])

df.createOrReplaceTempView("pessoas_temp")

# CREATE OR REPLACE é o mais seguro — funciona na 1ª e nas seguintes execuções
spark.sql("""
    CREATE OR REPLACE TABLE ceph_cat.default.pessoas_iceberg
    USING iceberg
    AS SELECT * FROM pessoas_temp
""")

spark.sql("SELECT * FROM ceph_cat.default.pessoas_iceberg").show()
```
